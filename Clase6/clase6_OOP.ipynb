#### 3. Constructor \__init__\, atributos y 'self'

el método \__init__ se ejecuta automáticamente cuando se crea una instancia.

Su función es inicializar el estado del objeto.

El parámetro *self* representa al objeto actual.

```python
# Ejm:
class PersonaEncuestada:
    pais = "Perú"
    
    def __init__(self, dni, nombre, edad, departamento):
        self.dni = dni
        self.nombre = nombre
        self.edad = edad
        self.departamento = departamento
        self.respuestas = []
```

En la línea:

self.nombre = nombre

- *self.nombre* es el atributo almacenado dentro del objeto

Los *parámetros* existen durante la ejecución del método.
Los *atributos* permanecen asociados al objeto.

```python
# Ejm:
persona_1 = PersonaEncuestada(
    dni = "60504536",
    nombre = "Roy Solier",
    edad = 24,
    departamento = "Ayacucho"
)

persona_2 = PersonaEncuestada(
    dni = "74088451",
    nombre = "Maria Rufino",
    edad = 26,
    departamento = "Huanuco"
)
```

#### 4. Métodos

Un **método** es una función definida dentro de una clase.

Los métodos representan comportamientos del objeto.

En esta versión agregamos métodos para:
- registrar respuestas;
- contar respuestas;
- modificar el departamento;
- mostrar un resumen.

```python
# Ejm:
class PersonaEncuestada:
    pais = "Perú"
    
    def __init__(self, dni, nombre, edad, departamento):
        self.dni = dni
        self.nombre = nombre
        self.edad = edad
        self.departamento = departamento
        self.respuestas = []
        
    def agregar_respuestas(self, respuesta):
        self.respuestas.append(respuesta)
        
    def cantidad_respuestas(self):
        return len(self.respuestas)
    
    def cambiar_departamento(self, nuevo_departamento):
        self.departamento = nuevo_departamento
        
    def mostrar_resumen(self):
        return (
            f"{self.nombre} tiene "
            f"{self.cantidad_respuestas()} respuestas registradas"
        )
```

```python
# Agredar datos:
persona_1 = PersonaEncuestada(
    dni = "60504536",
    nombre = "Roy Solier",
    edad = 24,
    departamento = "Ayacucho"
)

# Agregar respuesta
persona_1.agregar_respuestas("Sí")
persona_1.agregar_respuestas("No")

print(persona_1.respuestas)
print(persona_1.cantidad_respuestas())
print(persona_1.mostrar_resumen())
```

Cuando escribimos:
- *persona_1.agregar_respuestas("Sí")*

python lo interpreta conceptualmente:

- *PersonaEncuestada.agregar_respuesta(persona_1, "Sí")*

Per eso *self* representa a *persona_1* surante esa llamada. 

```python
print(persona_1.departamento)
persona_1.cambiar_departamento("Lima")
print(persona_1.departamento)
```

#### *print* y *return* no significan lo mismo
- *print()* muestra información en pantalla.
- *return* devueleve un valor al lugar desde donde se llamó al método.

Un valor devuelto puede almacenarse, reutilizarse o cambiarse con otras operaciones.

```python
# Ejm:
resumen_guardado = persona_1.mostrar_resumen()

print("Valor guardado:")
print(resumen_guardado)
```

#### 5. Validación y estado válido

Una clase no solo almacena datos. También puede proteger las reglas del problema.

Aplicaremos estas reglas:
- El DNI debe contener exactamente 8 dígitos.
- La edad debe estar entre 0 y 120.
- La respuesta solo puede ser: "Sí", "No" o "No Sabe".
- El nombre y el departamento no pueden estar vacíos.

```python
# Ejm:
class PersonaEncuestada:
    pais = "Perú"
    RESPUESTAS_VALIDAS = ("Sí", "No", "No sabe")
    
    def __init__(self, dni, nombre, edad, departamento):
        self._validar_dni(dni)
        self._validar_texto(nombre, "nombre")
        self._validar_edad(edad)
        self._validar_texto(departamento, "departamento")
        
        self.dni = dni
        self.nombre = nombre
        self.edad = edad
        self.departamento = departamento
        self.respuestas = []
        
    @staticmethod
    def _validar_dni(dni):
        if not isinstance(dni, str):
            raise TypeError("El DNI debe proporcionarse como texto.")
        if not dni.isdigit() or len(dni) != 8:
            raise ValueError("El DNI debe contener exactamente 8 digítos.")
    
    @staticmethod
    def _validar_texto(valor, nombre_campo):
        if not isinstance(valor, str):
            raise ValueError(f"El campo {nombre_campo} debe ser texto.")
        if not valor.strip():
            raise ValueError(f"El campo {nombre_campo} no puede estar vacío.")
    
    @staticmethod
    def _validar_edad(edad):
        if not isinstance(edad, int):
            raise TypeError("La edad debe ser un número entero.")
        if edad < 0 or edad > 120:
            raise ValueError("La edad debe estar entre 0 y 120.")
        
    def agregar_respuesta(self, respuesta):
        if respuesta not in self.RESPUESTAS_VALIDAS:
            opciones = ", ".join(self.RESPUESTAS_VALIDAS)
            raise ValueError(
                f"Respuesta no válida. Use una de estas opciones: {opciones}."
            )
        self.respuestas.append(respuesta)
        
    def cantidad_respuestas(self):
        return len(self.respuestas)
    
    def cambiar_departamento(self, nuevo_departamento):
        self._validar_texto(nuevo_departamento, "departamento")
        self.departamento = nuevo_departamento
    
    def cambiar_edad(self, nueva_edad):
        self._validar_edad(nueva_edad)
        self.edad = nueva_edad
        
    def mostrar_resumen(self):
        return (
            f"{self.nombre} tiene "
            f"{self.cantidad_respuestas()} respuestas registradas."
        )
        
    def __repr__(self):
        return (
            f"PersonaEncuestada(dni='{self.dni}', "
            f"nombre='{self.nombre}', edad='{self.edad}', "
            f"departamento='{self.departamento}')"
        )
        
```

#### 7. Composición

La composición ocurre cuando un objeto contiene otros objetos.

En nuestro caso:
- Una encuesta tiene participantes.

Una encuesta no es una persona. Por eso no corresponde utilizar herencia.

Creamos una clase 'Encuesta' que almacenará objetos 'PersonaEncuesta'.

```python
# Ejm:
class Encuesta:
    def __init__(self, nombre):
        if not isinstance(nombre, str) or not nombre.strip():
            raise ValueError("El nombre de la encuesta no puede estar vacío.")
        
        self.nombre = nombre
        self.participantes = []
        
    # tener mucho cuidado: persona es un objeto:
    def agregar_participante(self, persona):
        # persona es un objeto de tipo PersonaEncuestada (class PersonaEncuestada)
        if not isinstance(persona, PersonaEncuestada):
            raise TypeError(
                "Solo se puede agregar objeto de tipo PersonaEncuestada."
            )
        if self.buscar_por_dni(persona.dni) is not None:
            raise ValueError(
                f"Ya existe un participante con DNI {persona.dni}."
            )
        
        # l: list -> L.append(<cual tipo de elemento>)   
        self.participantes.append(persona)
        
    def buscar_por_dni(self, dni):
        for participante in self.participantes:
            if participante.dni == dni:
                return participante
        return None
    
    def cantidad_participantes(self):
        return len(self.participantes)
    
    def listar_nombres(self):
        # L = []
        # for participante in self.participantes:
        #    L.append(participante.nombre)
        # return L
        return [participante.nombre for participante in self.participantes]
    
    def total_respuestas(self):
        return sum(participante.cantidad_respuestas() for participante in self.participantes)
    
    def mostrar_resumen(self):
        lineas = [
            f"Encuesta: {self.nombre}",
            f"Cantidad de participantes: {self.cantidad_participantes()}",
            f"Total de respuestas: {self.total_respuestas()}"
        ]
        
        for participante in self.participantes:
            lineas.append(
                f"- {participante.nombre}: "
                f"{participante.cantidad_respuestas()} respuestas"
            )
            
        return "\n".join(lineas)
```

```python
# Creo el objeto encuesta (instanciación de la clase Encuesta)
encuesta = Encuesta("Encuesta Nacional 2026")

encuesta
```

#### 8. Herencia simple

La herencia permite crear una clase especializada a partir de otra clase más general.

Crearemos dos subclases:
- RepresentanteHogar
- RepresentanteEmpresa

Ambas son tipos de PersonaEncuestada

```python
class RepresentanteHogar(PersonaEncuestada):
    def __init__(self, dni, nombre, edad, departamento, codigo_vivienda):
        super().__init__(dni, nombre, edad, departamento)
        
        if not isinstance(codigo_vivienda, str) or not codigo_vivienda.strip():
            raise ValueError("El código no puede estar vacío")
        
        self.codigo_vivienda = codigo_vivienda
        
    def mostrar_tipo(self):
        return "Representante de hogar"
    
    def mostrar_detalle(self):
        return (
            f"{self.nombre} representante de vivienda "
            f"{self.codigo_vivienda}."
        )
        
class RepresentanteEmpresa(PersonaEncuestada):
    def __init__(self, dni, nombre, edad, departamento, empresa, cargo):
        super().__init__(dni, nombre, edad, departamento)
        
        self._validar_texto(empresa, "empresa")
        self._validar_texto(cargo, "cargo")
        
        self.empresa = empresa
        self.cargo = cargo
        
    def mostrar_tipo(self):
        return "Representante de empresa"
    
    def mostrar_detalle(self):
        return (
            f"{self.nombre} representa a {self.empresa} "
            f" como {self.cargo}"
        )
```

#### ¿Qué hace super()?

La expresión :
- super().__init__(dni, nombre, edad, departamento)

llama al método __init__ de la clase padre.

Así reutilizamos:
- validación del DNI;
- validación del nombre;
- validación de la edad;
- validación del departamento;
- creación de la lista de respuestas.


```python
representante_hogar = RepresentanteHogar(
    dni = "70123456",
    nombre = "Rosa Huamán",
    edad = 37,
    departamento = "Junín",
    codigo_vivienda = "VIV-001"
)

representante_empresa = RepresentanteEmpresa(
    dni = "45123789",
    nombre = "Lucia Ramos",
    edad = 41,
    departamento = "Arequipa",
    empresa = "Empresa Andina",
    cargo = "Gerente"
)

print(representante_hogar.mostrar_detalle())
print(representante_empresa.mostrar_detalle())
```

```python
print(isinstance(representante_hogar, RepresentanteHogar), type(representante_hogar))
print(isinstance(representante_hogar, PersonaEncuestada), type(representante_hogar))

print(isinstance(representante_empresa, RepresentanteEmpresa), type(representante_empresa))
```

El comando **isinstance()**, sirve para verificar si un objeto o variable pertenece a un tipo de dato específico

### 11. Encapsulación

La encapsulación busca controlar el acceso y la modificación del estado interno del objeto.

En Python:
- _atributo indica por convención que el atributo es de uso interno.
- __atributo activa *name mangling*.
- Ninguno representa seguridad absoluta.

Crearemos una clase que controle la modificación de la edad.

```python
# Ejm: Crea un atributo de clase privada llamada __age:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age  # Atributo privado
        
p1 = Person("Roy", 24)
print(p1.name)
#print(p1.__age)  # Esto provocará un error

# No se puede acceder a atributos privadas directamente desde fuera de la clase
```

Atributo privada

```python
# Para acceder a una propiedad privada, puede crear un método getter:
# Ejm: utilice un método getter para acceder a una propiedad privada:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age
        
    def get_age(self):
        return self.__age
    
p1 = Person("Roy", 24)
print(p1.get_age())
```

```python
# para modificar un atributo privada, puede crear un método setter.
# El método setter también puede validar el valor antes de configurarlo:
# Ejm: Utilice un método setter para cambiar una propiedad privada:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age
    
    def get_age(self):
        return self.__age
    
    def set_age(self, new_age):
        if new_age > 0:
            self.__age = new_age
        else:
            print("Age must be positive")
            
p1 = Person("Roy", 24)
print(p1.get_age())

p1.set_age(25)
print(p1.get_age())
            
```

```python
# Ejm: utilice la encapsulación para proteger y validar datos:
class Student:
    def __init__(self, name):
        self.name = name
        self.__grade = 0
        
    def set_grade(self, new_grade):
        if 0 <= new_grade <= 100:
            self.__grade = new_grade
        else:
            print("grade must be between 0 and 100")
            
    def get_grade(self):
        return self.__grade
    
    def get_status(self):
        if self.__grade >= 60:
            return "Passed"
        else:
            return "Failed"
        
student = Student("Roy Solier")

student.set_grade(85)

print(student.get_grade())
print(student.get_status())
```

Métodos privados

```python
# También puedes hacer que los métodos sean privadas usando el prefijo de doble guión bajo:
# Ejm: Crea un método:
class Calculator:
    def __init__(self):
        self.result = 0
        
    def __validate(self, num):
        if not isinstance(num, (int, float)):
            return False
        return True
    
    def add(self, num):
        if self.__validate(num):
            self.result += num
        else:
            print("Invalid number")

calc = Calculator()
calc.add(10)
calc.add(5)
print(calc.result)
# calc.__validate(5)  # This would cause an error
```

```python
# Ejm del profe clifor
class PersonaEncapsulada:
    def __init__(self, nombre, edad, codigo_interno):
        if not isinstance(nombre, str) or not nombre.strip():
            raise ValueError("El nombre no puede estar vacío.")
        
        self.nombre = nombre
        self._edad = None
        self.__codigo_interno = codigo_interno
        
        self.cambiar_edad(edad)
        
    def obtener_edad(self):
        return self._edad
    
    def cambiar_edad(self, nueva_edad):
        if not isinstance(nueva_edad, int):
            raise TypeError("La edad debe ser un número entero.")
        if nueva_edad < 0 or nueva_edad > 120:
            raise ValueError("La edad debe estar entre 0 y 120")
        
        self._edad = nueva_edad
        
    def verificar_codigo(self, codigo):
        return codigo == self.__codigo_interno
```

```python
persona_encapsulada = PersonaEncapsulada(
    nombre = "Maria López",
    edad = 32,
    codigo_interno = "ENC-2026-001"
)

print(persona_encapsulada.obtener_edad())

persona_encapsulada.cambiar_edad(32)

print(persona_encapsulada.obtener_edad())
print(persona_encapsulada.verificar_codigo("ENC-2026-001"))
print(persona_encapsulada.verificar_codigo("CODIGO-INCORRECTO"))
```

```python
try:
    persona_encapsulada.cambiar_edad(-10)
except ValueError as error:
    print(f"Error controlado: {error}")
```

### 12. Herencia múltiple y MRO

Python permite que una clase herede de más de una clase.

Sin embargo, esta característica aumenta la complejidad.

Cuando varias clases tienen métodos con el mismo nombre, Python utiliza el **Method Resolution Orden (MRO)** para determinar el orden de búsqueda.

```python
# Ejm:
class Registrable:
    def registrar(self):
        return "Registros realizados"
    
class Auditable:
    def auditar(self):
        return "Auditoría realizada"
    
class ParticipanteAdministrativo(Registrable, Auditable):
    pass

participante_administrativo = ParticipanteAdministrativo()

print(participante_administrativo.registrar())
print(participante_administrativo.auditar())
print([clase.__name__ for clase in ParticipanteAdministrativo.mro()])
```

### 13. Herencia multinivel

La herencia multinivel ocurre cuando una clase hereda de otra que ya heredaba de una clase anterior.

```python
# Ejm:
class Entidad:
    def identificar_nivel(self):
        return "Entidad"
    
class PersonaGeneral(Entidad):
    def identificar_persona(self):
        return "Persona"
    
class Encuestador(PersonaGeneral):
    def identificar_rol(self):
        return "Encuestador"
    
encuestador = Encuestador()

print(encuestador.identificar_nivel())
print(encuestador.identificar_persona())
print(encuestador.identificar_rol())
```

Cree una subclase llamada 'RepresentanteInstitucion que herede de PersonaEncuestada.

Debe agregar:

```python
class RepresentanteInstitucion(PersonaEncuestada):
    def __init__(self, dni, nombre, edad, departamento, institucion, cargo):
        super().__init__(dni, nombre, edad, departamento)
        
        self._validar_texto(institucion, "institucion")
        self._validar_texto(cargo, "cargo")
        
        self.institucion = institucion
        self.cargo = cargo
        
    def mostrar_tipo(self):
        return "Representante de institucion"
    
    def mostrar_detalle(self):
        return (
            f"{self.nombre} pertenece a la institucion "
            f"{self.institucion} y ocupa el cargo de {self.cargo}"
        )
```
