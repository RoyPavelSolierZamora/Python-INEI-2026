#### 1. Resolver el probelma sin POO

Antes de utilizar clases, representamos a una persona encuestada mediante un diccionario.

La persona tendrá:
- DNI
- nombre
- edad
- departamento
- respuestas registradas

```python
persona_diccionario = {
    "dni": "60504536",
    "nombre": "Roy Solier",
    "edad": 34,
    "departamento": "Lima",
    "respuestas": []
}

print(persona_diccionario)
```

```python
def agregar_respuesta_diccionario(persona, respuesta):
    persona["respuestas"].append(respuesta)
    
def cantidad_respuestas_diccionario(persona):
    return len(persona["respuestas"])

def mostrar_resumen_diccionario(persona):
    return (
        f"{persona['nombre']} tiene "
        f"{cantidad_respuestas_diccionario(persona)} respuestas registradas."
    )
```

```python
agregar_respuesta_diccionario(persona_diccionario, "Sí")
agregar_respuesta_diccionario(persona_diccionario, "No")
```

```python
print(persona_diccionario["respuestas"])
print(mostrar_resumen_diccionario(persona_diccionario))
```

la solución funciona, pero presenta algunas limitaciones.
- Los datos están separados de las funciones.
- Debvemos recordar qué las funciones corresponden a cada diccionario.
- Podemos escribir mal una clave.
- Es posible crear diccionarios incompletos.
- Noexixten validaciones automáticas.

```python
try:
    print(persona_diccionario["nombres"])
except KeyError as error:
    print(f"Error controlado: la clave {error} no existe.")
```

La programación Orientada a Objetos ayuda a reunir los datos y las operaciones relacionadas dentro de una misma estructura.

#### 2. Clase, objeto e instancia

##### **Clase**

Una **clase** es un modelo que define:
- qué datos tendrán los objetos;
- qué operaciones podrán realizar;
- qué reglas deberán cumplir.

##### **Objeto**

Un **objeto** es una entidad concreta creada a partir de una clase.

##### **Instancia**

Una **instancia** es un objeto perteneciente a una clase. En la práctica, "objeto"  e "instancia" suelen usarse como equivalentes.

```python
# Ejm:
class PersonaVacia:
    pass

persona_vacia = PersonaVacia()  # Se convierte en un objeto o instancia

print(type(persona_vacia))
print(isinstance(persona_vacia, PersonaVacia))
```

#### 3. Constructor \__init__\, atributos y 'self'

el método \__init__ se ejecuta automáticamente cuando se crea una instancia.

Su función es inicializar el estado del objeto.

El parámetro *self* representa al objeto actual.

```python
# Ejm:
class PersonaEncuestada:
    pais = "Perú"
    
    def __init__(self, dni, nombre, edad, departamento):
        self.dni = dni
        self.nombre = nombre
        self.edad = edad
        self.departamento = departamento
        self.respuestas = []
```

En la línea:

self.nombre = nombre
- *nombre* es el parámetro recibido.
- *self.nombre* es el atributo almacenado dentro del objeto

Los *parámetros* existen durante la ejecución del método.
Los *atributos* permanecen asociados al objeto.

```python
# Ejm:
persona_1 = PersonaEncuestada(
    dni = "60504536",
    nombre = "Roy Solier",
    edad = 24,
    departamento = "Ayacucho"
)

persona_2 = PersonaEncuestada(
    dni = "74088451",
    nombre = "Maria Rufino",
    edad = 26,
    departamento = "Huanuco"
)
```

```python
print(persona_1.nombre)
print(persona_2.nombre)
print(persona_1.respuestas)
print(persona_1 is persona_2)
```

Ambas personas pertenecen a la misma clase, pero son objetos diferentes y tienen estados independientes.

#### 4. Métodos

Un **método** es una función definida dentro de una clase.

Los métodos representan comportamientos del objeto.

En esta versión agregamos métodos para:
- registrar respuestas;
- contar respuestas;
- modificar el departamento;
- mostrar un resumen.

```python
# Ejm:
class PersonaEncuestada:
    pais = "Perú"
    
    def __init__(self, dni, nombre, edad, departamento):
        self.dni = dni
        self.nombre = nombre
        self.edad = edad
        self.departamento = departamento
        self.respuestas = []
        
    def agregar_respuestas(self, respuesta):
        self.respuestas.append(respuesta)
        
    def cantidad_respuestas(self):
        return len(self.respuestas)
    
    def cambiar_departamento(self, nuevo_departamento):
        self.departamento = nuevo_departamento
        
    def mostrar_resumen(self):
        return (
            f"{self.nombre} tiene "
            f"{self.cantidad_respuestas()} respuestas registradas"
        )
```

```python
# Agredar datos:
persona_1 = PersonaEncuestada(
    dni = "60504536",
    nombre = "Roy Solier",
    edad = 24,
    departamento = "Ayacucho"
)

# Agregar respuesta
persona_1.agregar_respuestas("Sí")
persona_1.agregar_respuestas("No")

print(persona_1.respuestas)
print(persona_1.cantidad_respuestas())
print(persona_1.mostrar_resumen())
```

Cuando escribimos:
- *persona_1.agregar_respuestas("Sí")*

python lo interpreta conceptualmente:

- *PersonaEncuestada.agregar_respuesta(persona_1, "Sí")*

Per eso *self* representa a *persona_1* surante esa llamada. 

```python
print(persona_1.departamento)
persona_1.cambiar_departamento("Lima")
print(persona_1.departamento)
```

#### *print* y *return* no significan lo mismo
- *print()* muestra información en pantalla.
- *return* devueleve un valor al lugar desde donde se llamó al método.

Un valor devuelto puede almacenarse, reutilizarse o cambiarse con otras operaciones.

```python
# Ejm:
resumen_guardado = persona_1.mostrar_resumen()

print("Valor guardado:")
print(resumen_guardado)
```

#### 5. Validación y estado válido

Una clase no solo almacena datos. También puede proteger las reglas del problema.

Aplicaremos estas reglas:
- El DNI debe contener exactamente 8 dígitos.
- La edad debe estar entre 0 y 120.
- La respuesta solo puede ser: "Sí", "No" o "No Sabe".
- El nombre y el departamento no pueden estar vacíos.

```python
# Ejm:
class PersonaEncuestada:
    pais = "Perú"
    RESPUESTAS_VALIDAS = ("Sí", "No", "No sabe")
    
    def __init__(self, dni, nombre, edad, departamento):
        self._validar_dni(dni)
        self._validar_texto(nombre, "nombre")
        self._validar_edad(edad)
        self._validar_texto(departamento, "departamento")
        
        self.dni = dni
        self.nombre = nombre
        self.edad = edad
        self.departamento = departamento
        self.respuestas = []
        
    @staticmethod
    def _validar_dni(dni):
        if not isinstance(dni, str):
            raise TypeError("El DNI debe proporcionarse como texto.")
        if not dni.isdigit() or len(dni) != 8:
            raise ValueError("El DNI debe contener exactamente 8 digítos.")
    
    @staticmethod
    def _validar_texto(valor, nombre_campo):
        if not isinstance(valor, str):
            raise ValueError(f"El campo {nombre_campo} debe ser texto.")
        if not valor.strip():
            raise ValueError(f"El campo {nombre_campo} no puede estar vacío.")
    
    @staticmethod
    def _validar_edad(edad):
        if not isinstance(edad, int):
            raise TypeError("La edad debe ser un número entero.")
        if edad < 0 or edad > 120:
            raise ValueError("La edad debe estar entre 0 y 120.")
        
    def agregar_respuesta(self, respuesta):
        if respuesta not in self.RESPUESTAS_VALIDAS:
            opciones = ", ".join(self.RESPUESTAS_VALIDAS)
            raise ValueError(
                f"Respuesta no válida. Use una de estas opciones: {opciones}."
            )
        self.respuestas.append(respuesta)
        
    def cantidad_respuestas(self):
        return len(self.respuestas)
    
    def cambiar_departamento(self, nuevo_departamento):
        self._validar_texto(nuevo_departamento, "departamento")
        self.departamento = nuevo_departamento
    
    def cambiar_edad(self, nueva_edad):
        self._validar_edad(nueva_edad)
        self.edad = nueva_edad
        
    def mostrar_resumen(self):
        return (
            f"{self.nombre} tiene "
            f"{self.cantidad_respuestas()} respuestas registradas."
        )
        
    def __repr__(self):
        return (
            f"PersonaEncuestada(dni='{self.dni}', "
            f"nombre='{self.nombre}', edad='{self.edad}, "
            f"departamento='{self.departamento}')"
        )
        
```

#### 7. Composición

La composición ocurre cuando un objeto contiene otros objetos.

En nuestro caso:
- Una encuesta tiene participantes.

Una encuesta no es una persona. Por eso no corresponde utilizar herencia.

Creamos una clase 'Encuesta' que almacenará objetos 'PersonaEncuesta'.

```python
class Encuesta:
    def __init__(self, nombre):
        if not isinstance(nombre, str) or not nombre.strip():
            raise ValueError("El nombre de la encuesta no puede estar vacío.")
        
        self.nombre = nombre
        self.participantes = []
        
    # Tener mucho cuidado: persona es un objeto
    def agregar_participante(self, persona):
        # persona es un objeto de tipo PersonaEncuestada (class PersonaEncuestada)
        if not isinstance(persona, PersonaEncuestada):
            raise TypeError("Solo se pueden agrgar objetos de tipo PersonaEncuestada.")
        
        if self.buscar_por_dni(persona.dni) is not None:
            raise ValueError(f"Ya existe un participante con DNI {persona.dni}.")
        
        # l: list --> l.append(<cual tipo de elemento>)
        self.participantes.append(persona)
    
    def buscar_por_dni(self, dni):
        for participante in self.participantes:
            if participante.dni == dni:
                return participante
        return None
    
    def cantidad_participantes(self):
        return len(self.participantes)
    
    def listar_nombres(self):
        # L = []
        # for participante in self.participantes:
        #     L.append(participante.nombre)
        # return L
        return [participante.nombre for participante in self.participantes]
    
    def total_respuestas(self):
        return sum(
            participante.cantidad_respuestas()
            for participante in self.participantes
        )
        
    def mostrar_resumen(self):
        lineas = [
            f"Encuesta: {self.nombre}",
            f"Cantidad de participantes: {self.cantidad_participantes()}",
            f"Total de respuestas: {self.total_respuestas()}"
        ]
        
        for participante in self.participantes:
            lineas.append(
                f"- {participante.nombre}: "
                f"{participante.cantidad_respuestas()} respuestas"
            )
        
        return "\n".join(lineas)
```
