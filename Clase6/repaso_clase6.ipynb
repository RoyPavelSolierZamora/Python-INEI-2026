### 3. Constructor '--init--', atributos y 'self'

El método --init-- se ejecuta automáticamente cuando se crea una instancia.

Su función es inicializar el estado del objeto.

El parámetro 'self' representa al objeto actual.

```python
class PersonaEncuestada:
    pais = "Perú"
    
    def __init__(self, dni, nombre, edad, departamento):
        self.dni = dni
        self.nombre = nombre
        self.edad = edad
        self.departamento = departamento
        self.respuestas = []
```

En la línea:

self.nombre = nombre

- 'self.nombre' es el atributo almacenado dentro del objeto.

Los parámetros existen durante la ejecución del método.

Los atributos permanecen asociados al objeto.


```python
persona_1 = PersonaEncuestada(
    dni = "60504536",
    nombre = "Roy Solier",
    edad = 24,
    departamento = "Ayacucho"
)

persona_2 = PersonaEncuestada(
    dni = "74088451",
    nombre = "María Rufino",
    edad = 26,
    departamento = "Huánuco"
)

print(persona_1.nombre)
print(persona_2.nombre)
print(persona_1.respuestas)
print(persona_1 is persona_2)
```

### 4. Métodos

Un **método** es una función definida dentro de una clase.

Los métodos representan comportamientos del objeto.

En esta versión agregaremos métodos para:
- registrar respuestas;
- contar respuestas;
- modificar el departamento;
- mostrar un resumen.

```python
class PersonaEncuestada:
    pais = "Perú"
    
    def __init__(self, dni, nombre, edad, departamento):
        self.dni = dni
        self.nombre = nombre
        self.edad = edad
        self.departamento = departamento
        self.respuestas = []
    
    def agregar_respuesta(self, new_respuesta):
        self.respuestas.append(new_respuesta)
    
    def cantidad_respuestas(self):
        return (len(self.respuestas))
    
    def cambiar_departamento(self, new_depart):
        self.departamento = new_depart
    
    def mostrar_resumen(self):
        return (
            f"{self.nombre} tiene "
            f"{self.cantidad_respuestas()} respuestas registradas."
        )
```

```python
persona_1 = PersonaEncuestada(
    dni = "60504536",
    nombre = "Roy Solier",
    edad = 34,
    departamento = "Ayacucho"
)

persona_1.agregar_respuesta("Sí")
persona_1.agregar_respuesta("No")

print(persona_1.respuestas)
print(persona_1.cantidad_respuestas())
print(persona_1.mostrar_resumen())
```

Cuando escribimos:
- persona_1.agregar_respuesta("Sí")

Pyhton lo interpreta conceptualmente como:
PersonaEncuestada.agregar_respusta(persona_1, "Sí")

Por eso self representa a persona_1 durante esa llamada.

```python
print(persona_1.departamento)

persona_1.cambiar_departamento("Lima")

print(persona_1.departamento)
```

'print' y 'return' no significa lo mismo
- print() : muestra información en pantalla.
- return : devuelve un valor al lugar desde donde se llamó al método.

Un valor devuelto pude almacenarse, reutilizarse o combinarse con otras operaciones.

```python
resumen_guardado = persona_1.mostrar_resumen()

print("valor guardado:")
print(resumen_guardado)
```

### 5. Validación y estado válido

Una clase no solo almacena datos. También puede proteger las reglas del problema.

Aplicaremos estas reglas:
- El DNI debe contener exactamente 8 dígitos.
- La edad debe estar entre 0 y 120.
- La respuesta solo puede ser: 'Sí', 'No' o 'No sabe'.
- El nombre y el departamento no pueden estar vacíos.

```python
class PersonaEncuestada:
    pais = "Perú"
    RESPUESTAS_VALIDAS = ("Sí", "No", "No sabe")
    
    def __init__(self, dni, nombre, edad, departamento):
        self._validar_dni(dni)
        self._validar_texto(nombre, "nombre")
        self._validar_edad(edad)
        self._validar_texto(departamento, "departamento")
        
        self.dni = dni
        self.nombre = nombre
        self.edad = edad
        self.departamento = departamento
        self.respuestas = []
    
    @staticmethod
    def _validar_dni(dni):
        if not isinstance(dni, str):
            raise TypeError("El DNI debe proporcionarse como texto.")
        if not dni.isdigit() or len(dni) !=8:
            raise ValueError("El DNI debe contener exactamente 8 dígitos.")
    
    @staticmethod
    def _validar_texto(valor, nombre_campo):
        if not isinstance(valor, str):
            raise TypeError(f"El campo {nombre_campo} debe ser texto.")
        if not valor.strip():
            raise ValueError(f"El campo {nombre_campo} no puede estar vacío.")
    
    @staticmethod
    def _validar_edad(edad):
        if not isinstance(edad, int):
            raise TypeError("La edad debe ser un número entero.")
        if edad < 0 or edad > 120:
            raise ValueError("La edad debe estar entre 0 y 120.")
    
    def agregar_respuesta(self, new_respuesta):
        if new_respuesta not in self.RESPUESTAS_VALIDAS:
            opciones = ", ".join(self.RESPUESTAS_VALIDAS)
            raise ValueError(
                f"Respuesta no válida. Use una de estas opciones: {opciones}."
            )
        self.respuestas.append(new_respuesta)
    
    def cantidad_respuestas(self):
        return len(self.respuestas)
    
    def cambiar_departamento(self, new_depart):
        self._validar_texto(new_depart, "departamento")
        self.departamento = new_depart
    
    def cambiar_edad(self, new_edad):
        self._validad_edad(new_edad)
        self.edad = new_edad
    
    def mostrar_resumen(self):
        return (
            f"{self.nombre} tiene "
            f"{self.cantidad_respuestas()} respuestas registradas."
        )
    
    def __repr__(self):
        return (
            f"PersonaEncuestada(dni='{self.dni}', "
            f"nombre='{self.nombre}', edad='{self.edad}', "
            f"departamento='{self.departamento}')"
        )
```

```python
persona_1 = PersonaEncuestada(
    dni = "60504536",
    nombre = "Roy Solier",
    edad = 24,
    departamento = "Lima"
)

persona_1.agregar_respuesta("Sí")
persona_1.agregar_respuesta("No sabe")

print(persona_1)
print(persona_1.mostrar_resumen())
```

### 7. Composición

La composición ocurre cuando un objeto contiene otros objetos.

En nuestro caso:
- Una encuesta tiene participantes.

Una encuesta no es una persona. Por eso no corresponde utilizar herencia.

Crearemos una clase 'Encuesta' que almacenará objetos 'PersonaEncuesta'.

```python
class Encuesta:
    def __init__(self, nombre):
        if not isinstance(nombre, str) or not nombre.strip():
            raise ValueError("El nombre de la encuesta no puede estar vacío.")
        
        self.nombre = nombre
        self.participantes = []
        
    # Tener mucho cuidado: persona es un objeto
    def agregar_participante(self, persona):
        # persona es un objeto de tipo PersonaEncuestada (class PersonaEncuestada)
        if not isinstance(persona, PersonaEncuestada):
            raise TypeError(
                "Solo se pueden agregar objetos de tipo PersonaEncuestada."
            )
        if self.buscar_por_dni(persona.dni) is not None:
            raise ValueError(
                f"Ya existe un participante con DNI {persona.dni}."
            )
        # l: list -> l.append(<cual tipo de elemento>)
        self.participantes.append(persona)
    
    def buscar_por_dni(self, dni):
        for participante in self.participantes:
            if participante.dni == dni:
                return participante
        return None
    
    def cantidad_participantes(self):
        return len(self.participantes)
    
    def listar_nombres(self):
        return [participante.nombre for participante in self.participantes] #for comprimido
    
    def total_respuestas(self):
        return sum(
            participante.cantidad_respuestas()
            for participante in self.participantes
        )
        
    def mostrar_resumen(self):
        lineas = [
            f"Encuesta: {self.nombre}",
            f"Cantidad de participantes: {self.cantidad_participantes()}",
            f"Total de respuestas: {self.total_respuestas()}"
        ]
        for participante in self.participantes:
            lineas.append(
                f"- {participante.nombre}: "
                f"{participante.cantidad_respuestas()} respuestas"
            )
        
        return "\n".join(lineas)
```

### 8. Herencia simple

La herencia permite crear una clase especializada a partir de otra clase más general.

Crearemos dos subclases:
- RepresentanteHogar
- RepresentanteEmpresa

Ambas son tipos de PersonaEncuestada

```python
class RepresentanteHogar(PersonaEncuestada):
    def __init__(self, dni, nombre, edad, departamento, codigo_vivienda):
        super().__init__(dni, nombre, edad, departamento)
        
        if not isinstance(codigo_vivienda, str) or not codigo_vivienda.strip():
            raise ValueError("El código de vivienda no puede estar vacío.")
        
        self.codigo_vivienda = codigo_vivienda
        
    def mostrar_tipo(self):
        return "Representante de hogar"
    
    def mostrar_detalle(self):
        return (
            f"{self.nombre} representa la vivienda "
            f"{self.codigo_vivienda}."
        )

class RepresentanteEmpresa(PersonaEncuestada):
    def __init__(self, dni, nombre, edad, departamento, empresa, cargo):
        super().__init__(dni, nombre, edad, departamento)
        
        self._validar_texto(empresa, "empresa")
        self._validar_texto(cargo, "cargo")
        
        self.empresa = empresa
        self.cargo = cargo
        
    def mostrar_tipo(self):
        return "Representante de empresa"
    
    def mostrar_detalle(self):
        return (
            f"{self.nombre} representa a {self.empresa} "
            f"como {self.cargo}."
        )
```

#### ¿Qué hace 'super()'?

La expresión:
- super().--init--(dni, nombre, edad, departamento)

llama al método --init-- de la clase padre.

Así reutilizamos:
- validación del DNI;
- validación del nombre;
- validación de la edad;
- validación del departamento;
- creación de la lista de respuestas.


```python
representante_hogar = RepresentanteHogar(
    dni = "45823683",
    nombre = "Rosa Huamán",
    edad = 37,
    departamento = "Junín",
    codigo_vivienda = "VIV-001"
)

representante_empresa = RepresentanteEmpresa(
    dni = "56383926",
    nombre = "Lucía Ramos",
    edad = 41,
    departamento = "Arequipa",
    empresa = "Empresa Andina",
    cargo = "Gerente"
)

print(representante_hogar.mostrar_detalle())
print(representante_empresa.mostrar_detalle())
```

```python
print(isinstance(representante_hogar, RepresentanteHogar), type(representante_hogar))
print(isinstance(representante_hogar, PersonaEncuestada), type(representante_hogar))

print(isinstance(representante_empresa, RepresentanteEmpresa), type(representante_empresa))
```

### 9. Polimorfismo

El polimorfismo permite utilizar el mismo nombre de método en diferentes clases.

Cada clase puede reponder de manera distinta.

Ambas subclases implementan:
- mostra_tipo()
- mostrar_detalle()

pero cada una devuelve un resultado diferente.

```python
participantes_especializados = [
    representante_hogar,
    representante_empresa
]

for participante in participantes_especializados:
    print(participante.mostrar_tipo())
    print(participante.mostrar_detalle())
    print("---")
```

La instrucción es la misma:
- participante.mostrar_tipo()

pero el comportamiento depende de la clase real del objeto.

### 10. Sobreescritura de métodos

Una subclase puede reemplazar un método heredado.

Agregaremos un método 'mostrar_tipo()' general a una nueva clase base y luego lo sobreescribimos.

```python
class ParticipanteBase:
    def mostrar_tipo(self):
        return "Participante general"
    
class ParticipanteHogarDemo(ParticipanteBase):
    def mostrar_tipo(self):
        return "Participante de hogar"

class ParticipanteEmpresaDemo(ParticipanteBase):
    def mostrar_tipo(self):
        return "Participante de empresa"

objetos_demo = [
    ParticipanteBase(),
    ParticipanteHogarDemo(),
    ParticipanteEmpresaDemo()
]

for objeto in objetos_demo:
    print(objeto.mostrar_tipo())  # polimorfosis
```

### 11. Encapsulación

La encapsulación busca controlar el acceso y la modificación del estado interno del objeto.

En Python:
- _atributo indica por conversión que el atributo es de uso interno.
- __atributo *activa name mangling*
- Ninguno representa seguridad absoluta.

Crearemos una clase que controle la modificación de la edad.

```python
class PersonaEncapsulada:
    def __init__(self, nombre, edad, codigo_interno):
        if not isinstance(nombre, str) or not nombre.strip():
            raise ValueError("El nombre no puede estar vacío.")
        
        self.nombre = nombre
        self.edad = None
        self.__codigo_interno = codigo_interno
        
        self.cambiar_edad(edad)
    
    def obtener_edad(self):
        return self.__edad
    
    def cambiar_edad(self, new_edad):
        if not isinstance(new_edad, int):
            raise TypeError("La edad debe ser un número entero.")
        if new_edad < 0 or new_edad > 120:
            raise ValueError("La edad debe estar entre 0 y 120.")
        
        self.__edad = new_edad
        
    def verificar_codigo(self, codigo):
        return codigo == self.__codigo_interno
```

```python
persona_encapsulada = PersonaEncapsulada(
    nombre = "María López",
    edad = 32,
    codigo_interno = "ENC-2026-001"
)

print(persona_encapsulada.obtener_edad())

persona_encapsulada.cambiar_edad(33)

print(persona_encapsulada.obtener_edad())
print(persona_encapsulada.verificar_codigo("ENC-2026-001"))
print(persona_encapsulada.verificar_codigo("CODIGO-INCORRECTO"))
```

```python
try:
    persona_encapsulada.cambiar_edad(-10)
except ValueError as error:
    print(f"Error controlado: {error}")
```

#### Name mangling

El atributo:
- self.__codigo_interno

no puede acceder normalmente como:
- persona_encapsulada.__codigo_interno

Python transforma internamente el nombre

Esto reduce el acceso accidental, pero no representa privacidad absoluta.

```python
try:
    print(persona_encapsulada.__codigo_interno)
except AttributeError as error:
    print(f"Error controlado: {error}")
```

### 12. Herencia múltiple y MRO

Python permite que una clase herede de más de una clase.

Sin embargo, esta característica aumenta la complejidad

Cuando varias clases tienen métodos con el mismo nombre, Python utiliza el **Method Resolution Order (MRO)** para determinar el orden de búsqueda.
