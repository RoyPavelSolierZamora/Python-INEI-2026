### RegEx

Es una expresión regular, es una secuencia de caracteres que forma un pátron de búsqueda.

```python
# Importando el módulo RegEx
import re
```

```python
# utilizando el módulo RegEx
# Ejm: Busca la cadena para ver si empieza con "El" y termina con "España":
txt = "La lluvia en España"
x = re.search("^La.*España$", txt)

if x:
    print("Sí! Tenemos una coencidencia.")
else:
    print("No hay coencidencias.")
```

Funciones de RegEx
- **findall:** devuelve una lista que contiene todas las coincidencias.
- **search:** devuelve un objeto Match si hay una coincidencia en cualquier parte de la cadena.
- **split:** Devuelve una lista donde la cadena se ha dividido en cada coincidencia.
- **sub:** Remplaza una o varias coincidencias con una cadena.

Metacarácteres
- **[] :** Un conjunto de caracteres. Ejm: "[a-m]"
- **\ :** Señala una secuencia especial (también puede usarse para escapar caracteres especiales). Ejm: "\d"
- **. :** Cualquier carácter (excepto el carácter de salto de línea). Ejm: "he..o"
- **^ :** Comienza con. Ejm: "^hello"
- signo dollar : Termina con. Ejm: "planet$"
- '* :' Cero o más ocurrencias. Ejm: "he.*o"
- **+ :** Una o más ocurrencias. Ejm: "he.+o"
- **? :** cero o una ocurrencia. Ejm: "he.?o"
- **{} :** exactamente el número de ocurrencias especificado. Ejm: "he.{2}o"
- **| :** Cualquiera de las dos opciones. Ejm: "falls|stays"
- **() :** Captura y agrupa.

```python
# Ejm: []
txt = "La lluvia en España"

x = re.findall("[a-m]", txt)  # Encuentra todos los caracteres en minúscula alfabeticamente entre a y m.
print(x)
```

```python
# Ejm: \
txt = "Serán 59 dolares"

x = re.findall("\d", txt)  # Encuestra todos los caracteres numéricos
print(x)
```

```python
# Ejm: .
txt = "hello plante"

x = re.findall("he..o", txt)  # Busca una secuencia que comience con "hola", seguida de dos (culaquer) caracter y una "o".
print(x)
```

```python
# Ejm: ^
txt = "hello planet"

x = re.findall("^hello", txt)  # revisa si el string inicia con 'hello'
if x:
    print("Yes!, the string starts with 'hello':", x)
else:
    print("no match")
```

```python
# Ejm: $
txt = "hello planet"

x = re.findall("planet$", txt)  # revisa si el string termina con 'planet'
if x:
    print("yes, the string ends with planet:", x)
else:
    print("No natch")
```

```python
# Ejm: *
txt = "hello plante"

x = re.findall("he.*o", txt)  # Busca una secuencia que comience con "he", seguida de 0 o más caracteres (cualquiera) y una "o".
print(x)
```

```python
# Ejm: +
txt = "hello plante"

x = re.findall("he.+o", txt)  # Busca una secuencia que comience con "he", seguida de 1 o más caracteres (cualquiera) y una "o".
print(x)
```

```python
# Ejm: ?
txt = "hello planet"

x = re.findall("he.?o", txt)  # Busca una secuencia que comience con "he", seguida de 0 o 1 (cualquier) carácter y una "o".
print(x)

# Esta vez no encontramos ninguna coincidencia, porque no había cero, ni uno, sino dos caracteres entre "he" y la "o".
```

```python
# Ejm: {}
txt = "hello planet"

x = re.findall("he.{2}o", txt)  # Busca una secuencia que comience con "he", seguida exactamente de 2 (cualesquiera) caracteres y una "o".
print(x)
```

```python
txt = "the rain in Spain falls mainly in the plain!"

x = re.findall("falls|stays", txt)  # Comprueba si la cadena contiene "falls" o "stays".
print(x)

if x:
    print("yes, there is at least one match!")
else:
    print("No match")
```
