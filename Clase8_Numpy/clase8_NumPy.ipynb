### 1. ¿Qué problema resuelve NumPy?

Supongamos que tenemos los ingresos mensuales de varias personas y necesitamos aumentar cada valor en 10%.

Con una lista de Python podemos resolverlo mediante un ciclo Esto es correcto y utiliza conocimientos ya estudiados.

```python
# Ejm:
ingresos = [1200, 1500, 1800, 2100, 2500]

ingresos_aumentados = []

for ingreso in ingresos:
    nuevo_ingreso = ingreso * 1.10
    ingresos_aumentados.append(nuevo_ingreso)
    
print("Ingresos orifinales:", ingresos)
print("ingresos con aumento", ingresos_aumentados)
```

No hay nada incorrecto en esta solución: Sin embargo, cuando trabajamos con cientos de miles o millones de valores, necesitamos una forma más directa y especializada.

NumPy permite aplicar la operación a todo el array de una sola vez.

```python
import numpy as np

ingresos_np = np.array([1200, 1500, 1800, 2100, 2500])
ingresos_aumentados_np = ingresos_np * 1.10

print("Array original:", ingresos_np)
print("Array con aumento", ingresos_aumentados_np)
```

#### ¿Qué observamos?

En NumPy no fue necesario un ciclo *for* para realizar el aumento.

A este modo de trabajo se le llama operación vectorizada.

#### ¿Qué es NumPy?

NumPy significa **Numeral Python**. Es una biblioteca orientada al manejo eficiente de array numéricos.

Su estructura principal es el objeto:
- ndarray

El término significa n_dimensional array, es decir, array de *n* dimensiones.

```python
print("Tipo de una lista de Python:", type(ingresos))
print("Tipo de un array Numpy:", type(ingresos_np) )
```

### 4. Propiedades principales de un array

Para comprender un array es necesario responder cuatro preguntas:
- ¿Cuántas dimensiones tiene?    -> *ndim*
- ¿Cuál es su forma?             -> *shape*
- ¿Cuántos elementos contiene?   -> *size*
- ¿Qué tipos de datos almacena?  -> *dtype*

```python
datos = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print("Arrays:")
print(datos)

print("\nNúmero de dimensiones:", datos.ndim)
print("Forma de array:", datos.shape)  # (2, 3): 2 filas, 3 columnas
print("Cantidad total de elementos:", datos.size)  # 2x3 = 6 elementos
print("Qué tipo de datos almacena:", datos.dtype)
```

Diferencia entre 'ndim', 'shape' y 'size':
- Propiedad   |       Pregunta que responde
- ndim       ->      ¿Cuántas dimensiones tiene?
- shape      ->      ¿Cuántas filas y columnas tiene?
- size       ->      ¿Cuántos elementos contiene en total?
- dtype      ->      ¿Qué tipo de datos almacena?

#### 4.1 un array utiliza un tipo común

En una lista de Python pueden coexistir diferentes tipos de datos.

En NumPy, los elementos del array normalmente se convierten en un tipo común

```python
mezcla_numerica = np.array([10, 20.5, 30])

print(mezcla_numerica)
print("tipo comun:", mezcla_numerica.dtype)
```

Aunque algunos valores fueron escritos como enteros , NumPy utiliza un tipo decimal para evitar perder información.

Esto es importante porque los arrays NumPy están diseñados para realizar cálculos uniformes y eficientes.


### 5. Acceso a los elementos de un arrreglo

NumPy utiliza índices, igual que las listas de Python.

El primer elemento se encuestra en la posición 0.

```python
# Ejm:
ventas = np.array([150, 175, 300, 250])

print("Primer elemento:", ventas[0])
print("Segundo elemento:", ventas[1])
print("Último elemento:", ventas[-1])
```

#### 5.1 Segmentación o slicing

La expresión:
- ventas[inicio:fin]
Selecciona desde la posición inicio hasta una posición anterior a fin.

```python
# Ejm:
print("Elementos desde la posición 1 hasta antes de la 4:", ventas[0:4])
print("primeros 3 elementos.", ventas[:3])
print("Desde la posición 2 hasta el final:", ventas[2:])
```

#### 5.2 Acceso en arrays de dos dimensiones

En un arreglo bidimensional usamos:
- matriz[fila, columna]
Los índices de filas y columnas también comienzan en cero.

```python
# Ejm:
matriz = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("Matriz completa:")
print(matriz)

print("\nElementos de la fila 0, columna 1:", matriz[0, 1])
print("Elementos de la fila 2, columna 2:", matriz[2, 2])
print("Primer fila:", matriz[0, :])
print("Segunda columna.", matriz[:, 1])
```

### 7. Filtrado mediante condiciones

Una de las condiciones más útiles de NumPy es seleccionar los elementos que cumplen una condición.

primero podemos evaluar una condición sobre todo el array

```python
# EJm:
ingresos = np.array([950, 1200, 1800, 2200, 3000])

condicion = ingresos > 1500

print("ingresos:", ingresos)
print("¿Es mayor que 1500?:", condicion)
```

Podemos utilizar directamente esa condición para extraer los valores.

```python
ingresos_mayores = ingresos[ingresos > 1500]

print("Ingresos mayores que 1500:", ingresos_mayores)
```

Este procedimiento se conoce como **filtrado booleano**

La expresión:

- ingresos > 1500

construye una máscara de valores verdaderos y falsos.

Luego:

- ingresos[ingresos > 1500]

devuelve únicamente los elementos cuyas posiciones tienen el valor True.

```python
# Ejm:
edades = np.array([12, 45, 23, 64, 53, 34, 23, 14, 18, 45, 37, 27])

edades[edades >= 18]

mayores_edad = edades[edades >= 18]
print("Personas de mayor edad:", mayores_edad)
print("\nDimensiones:", edades.ndim)
print("Cuántas filas y columnas:", edades.shape)
print("Elementos en total:", edades.size)
print("Tipo de datos almacenado:", edades.dtype)
```

```python
notas = np.array([12, 14, 15, 11, 9, 16, 20, 10, 17, 4])

notas_aprobadas = notas[notas >= 11]

print("Notas en egeneral:", notas)
print("Notas aprobadas:", notas_aprobadas)
print("\nDimensión:", notas.ndim)
print("Número de filas y columnas:", notas.shape)
print("Total de elementos:", notas.size)
print("Tipo de datos almacenados.", notas.dtype)
```

### 8. Operaciones vectorizadas

Una operación vectorizada se aplica a todos los elementos del array.

No necesitamos recorrer los valores mediante un ciclo explícito.

```python
# Ejm:
datos = np.array([10, 20, 30, 40])

print("Array original", datos)
print("Sumar 5:", datos + 5)
print("Multiplicar por 2:", datos * 2)
print("Dividir entre 10:", datos / 10)
print("Elevar al cuadrado:", datos ** 2)
```

#### 8.1 Diferencia entre una lista y array NumPy

La multiplicación tiene comportamientos distintos.

```python
# Ejm:
lista_python = [1, 2, 3]
array_numpy = np.array([1, 2, 3])

print("Lista multiplicada por 2:", lista_python * 2)
print("Array multiplicada por 2:", array_numpy * 2)
```

Interpretación:
- Una lista -> repite el contenido
- Un array -> multiplica cada elemento

Esta diferencioa es fundamental

#### 8.2 Operaciones entre arrays

También podemos operar dos arrays elemento por elemento.

Para ello deben tener formas compatibles.

```python
# Ejm:
ventas_enero = np.array([100, 150, 200, 250])
ventas_febrero = np.array([120, 140, 230, 270])

diferencia = ventas_febrero - ventas_enero
total_dos_meses = ventas_enero + ventas_febrero

print("Venta de enero:", ventas_enero)
print("ventas de febrero:", ventas_febrero)
print("Diferencia febrero - enero:", diferencia)
print("Total de los dos meses:", total_dos_meses)
```

```python
# Ejm:
bolsa_de_lima_1 = np.array([2.3, 5.2, 0.1, 1.1, 0.2])
bolsa_de_lima_2 = np.array([0.1, 1.9, 2.0, 3.3, 4.3])

diferencia = bolsa_de_lima_2 - bolsa_de_lima_1
total_dos_dias = bolsa_de_lima_1 + bolsa_de_lima_2

print("Bolsa de Lima día 1:", bolsa_de_lima_1)
print("Bolsa de Lima día 2.", bolsa_de_lima_2)
print("Diferencia día 2 y día 1:", diferencia)
print("Total de los dos días.", total_dos_dias)
```

### 10. Caso integrador: análisis de población distrital

Trabajaremos con un conjunto pequeñonde datos ficticios para aplicar los conceptos estudiados.

Tenemos cinco distritos y dos mediciones:
- población registrada;
- población proyectada.

El propósito es calcular:
- crecimiento absoluto;
- crecimiento porcentual;
- población promedio;
- distrito con mayor y menor población proyectada;
- distritos cuya población proyectada supera un límite.

```python
distritos = np.array([
    "Distrito Norte",
    "Distrito Sur",
    "Distrito Este",
    "Distrito Oeste",
    "Distrito Centro"
])

poblacion_registrada = np.array([120000, 80000, 150000, 95000, 200000])
poblacion_proyectada = np.array([126000, 84000, 162000, 98000, 218000])

print("Distritos:")
print(distritos)
print("\nPoblación regsitrada:")
print(poblacion_registrada)
print("\nPoblación proyectada:")
print(poblacion_proyectada)
```

##### 10.1 Crecimiento absoluto

El crecimiento absoluto se calcula restando:
- población proyectada - población registrada

Como ambos arrray tienen la misma cantidad de elementos, NumPy realiza la resta posición opor posición.

```python
crecimiento_absoluto = poblacion_proyectada - poblacion_registrada

print("Crecimiento absoluto por distrito:")
print(crecimiento_absoluto)
```

##### 10.2 Crecimiento porcentual

La fórmula es:
- ((proyectada - registrada) / registrada) x 100

```python
crecimiento_porcentual = (
    (poblacion_proyectada - poblacion_registrada)
    / poblacion_registrada
) * 100

print("Crecimiento porcentual:", crecimiento_porcentual)

# con dos decimáles
print("Crecimiento poblacional:", np.round(crecimiento_porcentual, 2))

#El arreglo original no se modifica; solamente se controla la forma en que se presenta los valores
```

##### 10.3 Promedio de población proyectada

```python
promedio_proyectado = np.mean(poblacion_proyectada)

print("Población proyectada promedio:", promedio_proyectado)
```

##### 10.4 Distrito con mayor población proyectada

*np.argmax()* devuelve la posición del valor máximo.

Después usamos esa posición para consultar el nombre del distrito y su población

```python
indice_mayor = np.argmax(poblacion_proyectada)

print("Posición del mayor valor:", indice_mayor)
print("Distrito con mayor población proyectada:", distritos[indice_mayor])
print("Población proyectada:", poblacion_proyectada[indice_mayor])
```

##### 10.5 Distrito con menor población proyectada

*np.argmin()* devuelve la posición del valor mínimo.

```python
indice_menor = np.argmin(poblacion_proyectada)

print("Posición del menor valor:", indice_menor)
print("Distrito con menor poblacion proyectada:", distritos[indice_menor])
print("Población proyectada:", poblacion_proyectada[indice_menor])
```

##### 10.6 Filtrar distritos con más de 100 000 habitantes proyectados

Primero se construye una condición booleano

```python
supera_limite = poblacion_proyectada >= 100000

print("Condición booleana:")
print(supera_limite)
```

Luego aplicamos la misma condición a los array relacionados.

Esto funciona porque 'distritos' y 'poblacion_proyectada' tienen la msima longitud y cada posición corresponde al mismo distrito.

```python
distritos_filtrados = distritos[supera_limite]
poblacion_filtrada = poblacion_proyectada[supera_limite]

print("Distritos con más de 100 000 habitantes proyectados:")
for distrito, poblacion in zip(distritos_filtrados, poblacion_filtrada):
    print(f"- {distrito}: {poblacion:,} habitantes")
```

NumPy realizó el filtrado y el ciclo se utilizó únicamente para presentar los resultados de una forma legíble.

##### 10.7 Presentación del resumen completo

```python
print("RESUMEN DEL ANÁLISIS")
print("-" * 55)

for i in range(len(distritos)):
    print(
        f"{distritos[i]:18} | "
        f"Registrada: {poblacion_registrada[i]:>7,} | "
        f"Proyectada: {poblacion_proyectada[i]:>7,} | "
        f"Crecimiento: {crecimiento_porcentual[i]:>5.2f}%"  
    )
    
print("-" * 55)
print(f"Promedio proyectado: {promedio_proyectado:,.2f}")
print(
    f"Mayor población: {distritos[indice_mayor]} "
    f"({poblacion_proyectada[indice_mayor]:,})"
)
print(
    f"Menor población: {distritos[indice_menor]}"
    f"({poblacion_proyectada[indice_menor]:,})"
)
```
