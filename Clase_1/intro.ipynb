Pregunta: ¿Por qué se comparan las letras? Es decir, por qué se puede realizar las operaciones relacionales?

Respuesta:
La razón por la que puedes usar operadores relacionales (>, <, ==, etc.) con texto es porque Python compara las cadenas utilizando un **orden lexicográfico** (básicamente, el orden alfabético de un diccionario) apoyándose en los valores numéricos exactos de cada letra.

Internamente, la computadora no lee las letras como texto; lee los números asignados a cada carácter por el estándar informático Unicode (o ASCII). Por ejemplo, la letra "a" minúscula siempre tiene el valor numérico 97, y la "n" tiene el valor 110. Las mayúsculas tienen valores diferentes (la "A" es 65).

```python
# Calcular la edad promedio de los encuestados
edades = []  # Creando una lista vacia

for personas in encuestados:
    edades.append(personas["edad"])

print(f" Las edades de las personas son: {edades}")
promedio = sum(edades) // len(edades)
print(f" El promedio de la edad es: {promedio}")
print(f" La edad mínima es: {min(edades)}")
print(f" La edad máxsima es: {max(edades)}")
```

```python
# Otra foma de printear:
edades = []  # Creando una lista vacia

for personas in encuestados:
    edades.append(personas["edad"])

promedio = sum(edades) // len(edades)

print(f"""
    Las edades de las personas son: {edades}
    El promedio de la edad es: {promedio}
    La edad mínima es: {min(edades)}
    La edad máxsima es: {max(edades)}
    """)
```

```python
# Puedo clasificar por grupo de edad:
menores = 0
adultos = 0
adultos_mayores = 0

for persona in encuestados:
    edad = persona["edad"]

    if edad < 18:
        menores += 1
    elif edad <= 65:
        adultos += 1
    else:
        adultos_mayores += 1

print("¿Cuántas personas de cada grupo hay?")
print("Menores de edad. ", menores)
print("Adultos: ", adultos)
print("Adultos mayores: ", adultos_mayores)
```

```python
# Puedo convertirlo en una salida tipo reporte
print(f"""
    REPORTE BÁSICO DE ENCUESTAS
    ===========================
    Total de encuestados: {len(encuestados)}
    Edad promedio: {round(promedio, 2)}
    Edad mínima: {min(edades)}
    Edad máxima: {max(edades)}

    Distribución por grupo etario:
    Menores de edad: {menores}
    Adultos: {adultos}
    Adultos mayores: {adultos_mayores}
      """)
```

```python
# Otra manera de ver este ejemplo:
edades = []
conteo_distritos = {}

menores = 0
adultos = 0
adultos_mayores = 0

for persona in encuestados:
    edad = persona["edad"]
    distrito = persona["distrito"]

    edades.append(edad)

    if edad < 18:
        menores += 1
    elif edad <= 65:
        adultos += 1
    else:
        adultos_mayores += 1

    if distrito in conteo_distritos:  # Significa: "Si el distrito ya existe como una 'llave' dentro de mi diccionario, entonces simplemente súmale 1 a la cantidad que ya tenía guardada (+= 1)".
        conteo_distritos[distrito] += 1  # "Toma lo que hay actualmente en conteo_distritos[distrito], súmale 1, y vuelve a guardar el resultado en ese mismo lugar". Es decir, hace exactamente esto:conteo_distritos[distrito] = conteo_distritos[distrito] + 1
    else:
        conteo_distritos[distrito] = 1  # Significa: "Si el distrito NO existe en mi diccionario (porque es la primera vez que lo veo), entonces créalo y asígnale el valor de 1 (= 1)".

print(f"edades = {edades}")
print(f"conteo_distritos = {conteo_distritos}")
```

```python
# Otro breve reporte
promedio = sum(edades) / len(edades)

print(f"""
REPORTE BÁSICO DE ENCUESTA
=============================
Total de encuestados: {len(encuestados)}
Edad promedio: {round(promedio)}
Edad mínima: {min(edades)}
Edad máxima: {max(edades)}

Distrito por grupo etario:
Menores de edad: {menores}
Adultos: {adultos}
Adultos mayores: {adultos_mayores}
    """
)
print("Encuestados por distritos:")
for distrito in conteo_distritos:
     print(distrito, ":", conteo_distritos[distrito])
```

2. Calculadora de presupuesto familiar: Haremos algo que es cercana a la vida diaria.

Objetivo: Dado un conjunto de ingresos y gastos, calcular saldo, porcentaje de ahorro y clasificar la situación financiera.

```python
ingresos = 2500
print(f"Ingresos = {ingresos}")

gastos = {
    "alquiler": 900,
    "comida": 650,
    "servicios": 220,
    "otros": 300
}

total_gastos = 0

for concepto in gastos:
    total_gastos += gastos[concepto]

print("Total gastos: ", total_gastos)

saldo = ingresos - total_gastos
print(f"Saldo = {saldo}")

porcentaje_ahorro = saldo / ingresos * 100
print(f"Porcentaje ahorrado = {porcentaje_ahorro}%")
```

```python
#Breve reporte
print(f"""
REPORTE DE GASTOS:
------------------
Ingresos: {ingresos}
Gastos: {total_gastos}
Saldo: {saldo}
Ahorro: {round(porcentaje_ahorro, 2)}
""")
```

```python
if saldo > 0:
    print("Situación: Superábit")
elif saldo == 0:
    print("Situación: Equilibrio")
else:
    print("Situación: Déficit")
```
