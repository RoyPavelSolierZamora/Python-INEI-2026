#### 3. Operador Ternario

El operador ternario le permite asignar un valor si una condición es verdadera y otro si es falsa:

Ejemplo: Asigne un valor a x:

Asigna el valor "¡FIN DE SEMANA!" si el número es superior a 5, en caso contrario "Día laborable":

```python
num = 6

x = "WEEKEND" if num > 5  else "Workday"

print(x)
```

Nota: El operador ternario no es un operador real, es un expresión condicional, o un declaración if abreviada.

```python
# En lugar de "Elif" se puede utilizar el operador ternario

"""Ejm: Asignar:
- "Viernes" si num es 5
- "Sábado" si num es 6
- "Sol" si num es 7
- de lo contrario asigne "día laborable":"""

num = 5

x = "Viernes" if num == 5 else "Sabádo" if num == 6 else "Domingo" if num == 7 else "Día laborable"

print(x)
```

#### 6. Comprensión de listas

La comprensión de listas ofrece una sintaxis más corta cuando desea crear una nueva lista basada en los valores de una lista existente.

Ejemplo:
Basado en una lista de frutas, desea una nueva lista que contenga solo las frutas con la letra "a" en el nombre.

```python
# Sin comprensión de listas tendrás que escribir una declaración for con una prueba condicional en su interior:
# Ejm: 
frutas = ["manzana", "plátano", "cereza", "kiwi", "mango"]
nueva_lista = []

for x in frutas:
    if "a" in x:
        nueva_lista.append(x)

print(nueva_lista)

# Con la comprensión de listas puedes hacer todo eso con una sola línea de código:
# Ejm:
frutass = ["manzana", "plátano", "cereza", "kiwi", "mango"]
nueva_listaa = [x for x in frutas if "a" in x]
print(nueva_listaa)
```

**La sintaxis:** newlist = [expression for item in iterable if condition == True]

El valor de retorno es una lista nueva, dejando la lista anterior sin cambios.

```python
# La condicón es como un filtro que solo acepta los elementos que evalúa True.
# Ejm:
frutas = ["manzana", "plátano", "cereza", "kiwi", "mango"]
nueva_list = [x for x in frutas if x != "manzana"]
print(nueva_list)
```

```python
# La condicional es opcional y se puede omitir
# Ejm: 
new_list = [x for x in frutas]
print(new_list)
```

```python
# El iterable puede ser cualquier objeto iterable, como una lista, tupla, conjunto, etc:
# Ejm: Utilizando la función range()
newlist = [x for x in range(10)]
print(newlist)

# Mismo ejm, pero con una condicion: Acepte sólo números inferiores a 5:

newlist = [ x for x in range(10) if x < 5]
print(newlist)
```

No puedes copiar una lista simplemente escribiendo list2 = list1, porque: list2 sólo será un referencia to list1, y los cambios realizados en list1 También se realizará automáticamente en list2.

#### 9. Unir a las listas

Hay varias formas de unir y concatenar dos o más listas

```python
# Operador + :
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(list3)

# Agregar todos los elementos de la lsita 2 a lista 1, una por una:
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

for x in list2:
    list1.append(x)

print(list1)

# Puedes usar el método extend(): método para agregar list2 al final de list1
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list1.extend(list2)
print(list1)
```

#### 5. Unir conjuntos

- Método union() y update(): Unen todos los elementos de ambos conjuntos.

```python

# Ejm con union(): Devuelve un nuevo set con todos los elementos de ambos sets.
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)
print(set3)

# También se cupede con |:
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set3 = set1 | set2
print(set3)

# Unir varios conjuntos con union():
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"Jhon", "Elena"}
set4 = {"apple", "banana", "cherry"}

myset = set1.union(set2, set3, set4)
print(myset)

# Utilizando |:
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"Jhon", "Elena"}
set4 = {"apple", "banana", "cherry"}

myset = set1 | set2 | set3 | set4
print(myset)

print("")
# Unir un conjunto y una tupla:
x = {"a", "b", "c"}
y = (1, 2, 3)

z = x.union(y)
print(z)
```

Nota: El  | El operador solo le permite unir conjuntos con conjuntos, y no con otros tipos de datos como usted lata con la  union() método.

- Método update(): Inserta todos los elementos de un conjunto en otro. Cambia el set original y no devuelve un nuevo set.

```python
# Ejm: El método inserta los elementos del set 2 en set 2.
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}

set1.update(set2)
print(set1)
```

- Método intersection(): conserva solo los duplicados de ambos sets, o los elementos que están presentes en ambos conjuntos.

```python
# Ejm: Unir set1 y set2, pero conservar solo los duplicados:
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.intersection(set2)
print(set3)

# También se puede usar el &:
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 & set2  # este método soloo permite unir entre sets y no como intersection() que une otros tipos.
print(set3)

print("")
# Método intersection_update(): Conserva solo lo duplicados, pero cambiará el set original en lugar de devolver un nuevo set.
# Ejm: mantega los elementos que existen en ambos set1 y set2:
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set1.intersection_update(set2)

print(set1)
print(set2)

print("")
# Ejm: Unir sets con valores True, False, 1 y 0, que también se consideran duplicados.
set1 = {"apple", 1, "banana", 0, "cherry"}
set2 = {False, "google", 1, "apple", 2, True}

set3 = set1.intersection(set2)
print(set3)
```

- Método difference(): Solo devolverá los elementos del primer set que no estén presentes en el otro set.

```python
# Ejm: mantenga todos los elementos del set 1 que no estén en el set 2.
set1 = {"apple","banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.difference(set2)
print(set3)

# También con el operador -:
set1 = {"apple","banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 - set2  # Este solo permite unir sets con set y no otros tipos de datos como difference().
print(set3)

print("")
# El método difference_update(): mantendrá los elementos del primer set que no esten en el otro set, pero cambiará el set original en lugar de devolver un nuevo set.
# Ejm: 
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.difference_update(set2)
print(set1)
```

Método symmetric_difference(): Diferencias simetrias, el método mantendrá solo los elementos que NO están presentes en ambos conjuntos.

```python
# Ejm:
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.symmetric_difference(set2)
print(set3)

# Otra forma: Solo para unir dos conjuntos
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1 ^ set2
print(set3)

print("")
# Método symmetric_difference_update(): El método también mantendrá la simetría, pero cambiará el conjunto original en lugar de devolver un nuevo conjunto.
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.symmetric_difference_update(set2)
print(set1)
```

### DICCIONARIO

se utilizan para almacenar valores de datos en parescon clave y valor (key, value).
