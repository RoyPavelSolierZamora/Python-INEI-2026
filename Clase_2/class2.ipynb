¿Por qué  todo string al convertir booleano es True? Excepto la cadena vacio "".

Esto se debe a un concepto de diseño en Python llamado **Truthy** (valores que actúan como verdaderos) y **Falsy** (valores que actúan como falsos).

Los creadores de Python decidieron que, para hacer la programación más intuitiva y rápida, la **"nada"** o el **"vacío"** siempre equivalen a la ausencia de datos **(False)**, mientras que la **existencia de algo** (cualquier contenido) equivale a **True**.

BUCLE: While y For

---> While

```python
i = 1

while i < 6:
    print(i)
    i += 1  # Recuerde incrementar i, de lo contrario el bucle continuará para siempre
```

La declaración de ruptuta "break"

```python
i = 1

while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
```

La declaración "continue"

```python
i = 0

while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

La declaración "else"

```python
i = 1

while i <  6:
    print(i)
    i += 1
else:
    print("i ya no es menor que 6")
```

```python
# El For se utiliza para iterar sobre una secuencia de una lista, tupla, diccionario, conjunto y texto.

fruits = ["apple", "banana", "cherry"]

for x in fruits:
    print(x)

# el bucle for no requiere que se configure una variable de indexación de antemano.
```

```python
# Ejm: Salga del bucle cuando x es "banana"

for x in fruits:
    print(x)
    if x =="banana":
        break
```

```python
# Ejm: Salga del bucle cuando x es "banana", pero esta vez la pausa viene de la impresión:

for x in fruits:
    if x == "banana":
        break
    print(x)
```

### FUNCIONES en Pyhton

Una función es un bloque de código que sólo se ejecuta cuando se llama

Como resultado, una función puede devolver datos.

Una función ayuda a evitar la repetición de código.

```python
# Creando una función:
'''
def nombreFunción():
'''

# Ejm:
def mi_funcion():
    print("Hola desde una función")

# llamando la función: (escribir su nombre de la función seguida del paréntesis)
mi_funcion()

# Se puede llamar a la misma función varias veces:
mi_funcion()
mi_funcion()
mi_funcion()

# Es una buena práctica utilizar nombres descriptivos que expliquen lo que hace la función.

```

```python
# Convertir la temperatura de Fahrenheit a Cdelsius si funciones

tem1 = 77
celsius1 = (tem1 - 32) * 5 / 9
print(celsius1)

tem2 = 95
celsius2 = (tem1 - 32) * 5 / 9
print(celsius2)

tem3 = 50
celsius3 = (tem3 - 32) * 5 / 9
print(celsius3)
```

```python
# Con funciones
def convert_tem(fahrenheit):
    return (fahrenheit - 32) * 5 / 9

convert_tem(77)

print(convert_tem(77))
print(convert_tem(95))
print(convert_tem(50))
```

```python
# Valores de parámetros predeterminados. Si se llama la función sin argumento utiliza este parámetro

def mi_definicion(parametro1 = "Amigo"):
    print("Hello", parametro1)

mi_definicion("Roy")
mi_definicion("argumento1")
mi_definicion()
```

Argumentos de palabra clave

```python
# Ejm: puedes enviar argumentos con el "clave = valor" sintaxis.

def mi_funcion(animal, nombre):
    print("Tengo un", animal)
    print("Mi", animal + " se llama", nombre)

mi_funcion(animal = "perro", nombre = "Layca")

# De esta manera, con los argumentos de palabras clave, el orden de los argumentos no importa.

mi_funcion(nombre = "Layca", animal = "perro")
```

```python
# Argumentos posicionales:  
# Ejm:

def mi_funcion(animal, nombre):
    print("Tengo un", animal)
    print("Mi", animal + " se llama", nombre)

mi_funcion("perro", "Layca")

# Cambiar el orden cambia el resultado.
mi_funcion("Layca", "perro")
```

```python
# Mezcla de argumentos posicionales y de palabras clave

def mi_funcion(animal, nombre, edad):
    print("Yo tengo un ", edad, "años", animal, "llamado", nombre)

mi_funcion("perro", nombre = "Layca", edad = 24)
```

```python
# Pasando diferentes tipos de datos (str, int, lista, dic, etc)

def mi_funcion(frutas):
    for fruta in frutas:
        print(fruta)

mis_frutas = ["apple", "banana", "cherry"]
mi_funcion(mis_frutas)

print(" ")

# Ejm: Enviar un diccionario como argumento:

def mi_funcion(persona):
    print("Nombre:", persona["nombre"])
    print("Edad:", persona["edad"])

mi_persona = {"nombre": "Roy", "edad": 24}
mi_funcion(mi_persona)
```
