## FUNCIONES

Una función es un bloque de código que solo se ejecuta cuando se llama

Como resultado, una función puede resolver datos

Una función ayuda a evitar la repetición de código

```python
# Ejm: creando una función:
def my_funcion():
    print("Hola desde la función")
```

```python
# Llamando la función - se puede llamar varias veces
my_funcion()
my_funcion()
my_funcion()
```

¿Por qué utilizar funcion?

```python
# Ejm: Sin funciones -código repetitivo
temp1 = 77
celsius1 = (temp1 - 32) * 5 / 9
print(celsius1)

temp2 = 92
celsius2 = (temp2 - 32) * 5 / 9
print(celsius2)

temp3 = 50
celsius3 = (temp3 - 32) * 5 / 9
print(celsius3)
```

```python
# Ejm: Con funciones - reutilizo
def conver_celsius(fahren):
    return (fahren - 32) * 5 / 9
    
print(conver_celsius(77))
print(conver_celsius(92))
print(conver_celsius(50))
```

Valores de retorno:

```python
# Ejm: Una función que devuelve un valor:
def get_greeting():
    return "Hola desde la función"

mensaje = get_greeting()
print(mensaje)

print(get_greeting())  # Valor utilizado directo
```

La declaración de aprobación ( pass):
- La definiciones de funciones no pueden estar vacías. si necesitas crear un marcador de posición de función sin ningún código, utilice la declaración *pass*

```python
# Ejm:
def mi_funcion():
    pass
```

### 1. Argumentos de funciones

La info se puede pasar a funciones como argumentos.

Los argumentos se especifícan después del nombre de la función, entre paréntesis.

```python
#Ejm: Una función con un argumento:
def mi_funcion(fname):
    print(fname + " Solier")
    
mi_funcion("Roy")
mi_funcion("Pavel")
```

Parametros Vs Argumentos
- Parámetro: es la variable que aparece entre paréntesis en la definición de la función.
- Argumento: es el valor real que se envía a la función cuando se llama. 

```python
# Ejm:
def mi_funcion(name):  # name is parameter
    print("Hello", name)
    
mi_funcion("Roy")  # "Roy" is an argument
```

```python
# la funciones pueden tener más de un argumento:
# Ejm: Esta función espera 2 argumentos y obtiene 2 argumentos:
def mi_funcion(fname, lname):
    print(fname, lname)
    
mi_funcion("Roy", "Solier")  # Si se intenta llamar a la función con un número incorrecto de argumentos, obtendra un error
```

Valores de parámetros predeterminados

```python
#Ejm:
def mi_funcion(name = "amigo"):
    print("Hola", name)
    
mi_funcion()
mi_funcion("Roy")
mi_funcion("Pavel")
```

```python
def mi_pais(pais = "Perú"):
    print("Soy de", pais)
    
mi_pais()
mi_pais("Brazil")
mi_pais(("EE.UU"))
```

Argumento nombrado o de palabra clave:
- clave = valor

```python
# Ejm:
def mi_funcion(animal, name):
    print("Tengo un", animal)
    print("Mi", animal, "se llama", name)
    
mi_funcion(animal = "Perro", name = "Layca")
mi_funcion(name = "Piriño", animal = "gato")
```

Argumentos posicionales:
- los argumentos posicionales deben estar en el orden correcto

```python
# Ejm:
def mi_funcion(animal, name):
    print("Tengo un", animal)
    print("My", animal, "Se llama", name)
    
mi_funcion("gato", "Piriño")  # Debe estar en orden correcto
mi_funcion("piriño", "gato")  # Cambiar el orden, cambia el resultado
```

Mezcla de argumentos posicionales y nombrados:

```python
# Ejm:
def mi_funcion(animal, name, age):
    print("tengo un", animal, "que tiene", age, "años", "y se llama", name)
    
mi_funcion("gato", name = "Piriño", age = 5)
```

Pasando diferentes tipos de datos:
- puede enviar cualquier tipo de datos como argumento a una función


```python
# Ejm: Enviar una lista como argumento:
def mi_funcion(frutas):
    for fruta in frutas:
        print(fruta)
        
mis_frutas = ["manzana", "plátano", "fresa"]
mi_funcion(mis_frutas)
```

```python
# Ejm: Enviar un diccionario como argumento:
def mi_funcion(persona):
    print("Nombre:", persona["nombre"])
    print("Edad:", persona["edad"])

mi_persona = {"nombre": "Roy", "edad": 24}
mi_funcion(mi_persona)
```

Valores de retorno:

```python
# Ejm:
def mi_funcion(x, y):
    return x + y

result = mi_funcion(5, 5)
print(result)
```

```python
# devolver diferentes tipos de datos:

# Ejm: Una función que devuelva una lista:
def mi_funcion():
    return ["manzana", "plátano", "fresa"]

frutas = mi_funcion()
print(frutas[0])
print(frutas[1])
print(frutas[2])
```

```python
# Ejm: Una función que devuelva una tupla:
def mi_funcion():
    return (10, 20)

x, y = mi_funcion()
print("x:", x)
print("y:", y)
```

Argumentos solo pocisionales:
- para especificar argumentos solo posicionales, se agrega ', /' después de los argumentos:

```python
# Ejm:
def mi_funcion(name, /):
    print("Hello", name)
    
mi_funcion("Roy")
#mi_funcion(name = Roy)  # Error
    
```

Argumentos que solo nombnrados o de palabra clave:
- para indicar ello, se agrega '*,' antes de los argumentos.

```python
# Ejm:
def mi_nombre(*, name):
    print("Hello", name)
    
mi_nombre(name = "Roy")
#mi_nombre("Roy")  # Error
```

Combinación de solo argumentos posicionales y palabras nombradas:
- /: son solos posicionales
- '*' : son solo nombrados

```python
def mi_funcion(a, b, /, *, c, d):
    return a + b + c + d

result = mi_funcion(5, 10, c = 15, d = 20)
print(result)
```

### 2. *args / **kwargs

Permite que las funciones acepten un número desconocido de argumentos.


#### 2.1 Argumentos arbitrarios - *args
- Agregue un * antes del nombre de parámetro
- *args permite que una función acepte cualquier número de argumentos posicionales
- dentro de la función, args se convierte en una tupla que contiene todos los argumentos posicionales
- de esa manera, la función recibirá una **tupla** de argumentos y podrá accederlos en secuencia.

```python
# Ejm: Usando *args para aceptar cualquier número de argumentos:
def mi_funcion(*kids):
    print("El niño más pequeño es", kids[2])
    
mi_funcion("Emil", "Tobias", "Linus")
```

```python
# Ejm: Accediendo a argumentos posicionales desde *args:
def mi_funcion(*args):
    print("Tipo:", type(args))
    print("Primer argumento:", args[0])
    print("Segundo argumento:", args[1])
    print("Todos los argumentos:", args)
    
mi_funcion("Emil", "Tobias", "Linus")
```

Uso de *args con argumentos regulares:

```python
# Ejm:
def mi_funcion(greeting, *names):
    for name in names:
        print(greeting, name)

mi_funcion("Hola", "Tobias", "Emil", "Linus")
```

```python
names = ("Roy", "Pavel", "Solier", "Zamora")

mi_funcion("Mi nombre completo es:", names)
```

```python
# *args es es útil cuando quieres crear funciones flexibles
# EJm : Una función que calcula la suma de cualquier número de valores:
def sumar(*numeros):
    total = 0
    for numero in numeros:
        total += numero
    return total

print(sumar(1, 2, 3))
```

```python
# Ejm Encontrar el valor máximo:
def mi_funcion(*numeros):
    return max(numeros)

print(mi_funcion(1, 2, 3, 5, 2, 8))
```

```python
def mi_funcion(*numeros):
    if len(numeros) == 0:
        return None
    max_num = numeros[0]
    for num in numeros:
        if num > max_num:
            max_num = num
    return max_num

print(mi_funcion(1, 2, 3, 5, 2, 8))
```

#### 2.2 Argumentos arbitrarios de los argumentos nombrados **kwargs
- agregue '**' antes del nombre del parámetro.
- **kwargs permite que una función acepte cualquier número de argumentos nombrados o palabra clave.
- dentro de la función, kwargs se convierte en un **diccionario** que contiene todos los argumentos de palabras nombrados.
- la función recibirá un diccionario de argumentos y podrá acceder a los elementos en consecuencia


```python
# Ejm: Usar **kwargs para aceptar cualquier número de argumentos de palabras clave:
def mi_funcion(**kids):
    print("Su apellido es:", kids["lname"])
    
mi_funcion(fname = "Roy", lname = "Solier")
```

```python
# Ejm: Accediendo a valores de **kwargs:
def mi_funcion(**myvar):
    print("Tipo:", type(myvar))
    print("Nombre:", myvar["nombre"])
    print("Edad:", myvar["edad"])
    print("Todo el dato:", myvar)

mi_funcion(nombre = "Roy", edad = 24)
```
