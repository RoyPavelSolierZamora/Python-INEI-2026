### 3. Mutabilidad, alias y copia de listas

**Idea clave:** Si hacemos 'b = a', ambas variables apuntan a la misma lista. Si se modifica la lista desde 'b', también se observa el cambio desde 'a'. Para evitarlo, use 'copy()' o slicing '[:]'.

**Problema 3.** Compare el comportamientode de una asignación directa y copia.

```python
original = [10, 20, 30]
alias = original
copia = original.copy()

alias.append(40)
copia.append(99)

print("original:", original)
print("alias:", alias)
print("copias:", copia)

# Explicación
  # se crea una lista llamada original
  # se crea una variable  alias homegenea a la original
  # se crea otra variable copias que captura los datos de la variable original pero que no se ve influenciada por ella
  # se agrega un número 49 a la variable alias y esto influye a la vez a la variable original
  # se crea un número 99 en la variable copias, y este no influye a la variable original
  
  # se imprime: [10, 20, 30, 40]
  # se imprime: [10, 20, 30, 40]
  # se imprime: [10, 20, 30, 99]
  
# Explicación breve:
  # alias y original son la misma lista.
  # copias es una lista distinta.
```

### 5 Comprensión de listas y diccionarios

**Idea clave:** Una comprensión permite construir una nueva colección en una sola expresión. Conviene usarla cuando la lógica es clara y breve.

**Problema 5.** Genere una lista de cuadrados para números impares y un diccionario número → cubo.

**Qué deben saber para una pregunta de desarrollo:** identificar entradas, proceso, salida esperada y explicar por qué el código cumple la regla. Si la pregunta pide analizar código, seguir línea por línea sin adivinar.


```python
numeros = [1, 2, 3, 4, 5, 6, 7]

cuadros_impares = [num*2 for num in numeros if num % 2 != 0]
cubos = {num: num** 3 for num in numeros}

print(cuadros_impares)
print(cubos)
```

### 6. Diccionarios para frecuencias y agrupamiento

**Idea clave:** Un diccionario permite asociar una clave con un valor. Es muy útil para contar, agrupar y consolidar información.

**Problema 6.** Dada una lista de sedes, cuente frecuencias y agrupe por primera letra.

**Qué deben saber para una pregunta de desarrollo:** identificar entradas, proceso, salida esperada y explicar por qué el código cumple la regla. Si la pregunta pide analizar código, seguir línea por línea sin adivinar.

```python
sedes = ["lima", "cusco", "lima", "piura", "ica", "cusco", "loreto"]

frecuencias = {}
for sede in sedes:
    frecuencias[sede] = frecuencias.get(sede, 0) + 1
    
print("Frecuencias.", frecuencias)

agrupado = {}
for sede in sedes:
    inicial = sede[0]
    agrupado.setdefault(inicial, []).append(sede)
    
print("Agrupado:", agrupado)
    
```

### 7. Bucles con continue, break y acumulación

**Idea clave:** 'continue' salta a la siguiente iteración. 'break' corta el bucle completo. En bucles anidados, para detener todo, suele convenir encapsular la búsqueda en una función y usar return.

**Problema 7.** Encuentre la primera combinación de dos listas cuya suma sea múltiplo de 5 y detenga la búsqueda.

**Qué deben saber para una pregunta de desarrollo:** identificar entradas, proceso, salida esperada y explicar por qué el código cumple la regla. Si la pregunta pide analizar código, seguir línea por línea sin adivinar.

```python
a = [2, 4, 6, 8]
b = [1, 3, 5, 7]

def primera_suma_multiplo_5(lista_a, lista_b):
    for x in lista_a:
        for y in lista_b:
            if (x + y) % 5 == 0:
                return x, y, x + y
    return None

resultado = primera_suma_multiplo_5(a, b)
print(resultado)

# return detiene la función completa, no solo el bucle interno
```

### 8. Alcance local, global y una alternativa preferible

**Idea clave:** Si una función asigna un valor a una variable, Python la considera local dentro de esa función. Usar 'global' modifica el estado externo, pero suele ser mejor evitarlo cuando se puede.

**Problema 8.** Corrija una función con contador usando global y luego una alternativa sin estado global.

**Qué deben saber para una pregunta de desarrollo:** identificar entradas, proceso, salida esperada y explicar por qué el código cumple la regla. Si la pregunta pide analizar código, seguir línea por línea sin adivinar.

```python
# Versión con global
contador_global = 0

def aumentar_global():
    global contador_global
    contador_global += 1
    return contador_global

print(aumentar_global())
print(aumentar_global())

# Versión preferible: pasar y retornar el nuevo estado
contador = 0

def aumentar(contador_actual):
    return contador_actual + 1

contador = aumentar(contador)
contador = aumentar(contador)
print("Sin global:", contador)
```

### 10. Funciones como parámetros y estrategia

**Idea clave:** En Python una función puede recibir otra función como parámetro. Esto permite cambiar el comportamiento sin reescribir la función principal.

**Problema 10.** Diseñe una función general que aplique una operación recibida como parámetro.

**Qué deben saber para una pregunta de desarrollo:** identificar entradas, proceso, salida esperada y explicar por qué el código cumple la regla. Si la pregunta pide analizar código, seguir línea por línea sin adivinar.

```python
def aplicar_operacion(a, b, operacion):
    return operacion(a, b)

def suma(a, b):
    return a + b

def mayor(a, b):
    return a if a > b else b

def multiplicar(a, b):
    return a * b

print(aplicar_operacion(8, 5, suma))
print(aplicar_operacion(8, 5, mayor))
print(aplicar_operacion(8, 5, multiplicar))
```
