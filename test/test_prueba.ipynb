### 6. Diccionarios para frecuencias y agrupamiento

**Idea clave:** Un diccionario permite asociar una clave con un valor. Es muy útil para contar, agrupar y consolidar información.

**Problema 6.** Dada una lista de sedes, cuente frecuencias y agrupe por primera letra.

```python
sedes = ["lima", "cusco", "lima", "piura", "ica", "cusco", "loreto"]

frecuencias = {}
for sede in sedes:
    frecuencias[sede] = frecuencias.get(sede, 0) + 1
    
print("Frecuencias.", frecuencias)

agrupado = {}
for sede in sedes:
    inicial = sede[0]
    agrupado.setdefault(inicial, []).append(sede)
    
print("Agrupado:", agrupado)
    
```

### 7. Bucles con continue, break y acumulación

**Idea clave:** 'continue' salta a la siguiente iteración. 'break' corta el bucle completo. En bucles anidados, para detener todo, suele convenir encapsular la búsqueda en una función y usar return.

**Problema 7.** Encuentre la primera combinación de dos listas cuya suma sea múltiplo de 5 y detenga la búsqueda.

**Qué deben saber para una pregunta de desarrollo:** identificar entradas, proceso, salida esperada y explicar por qué el código cumple la regla. Si la pregunta pide analizar código, seguir línea por línea sin adivinar.

```python
a = [2, 4, 6, 8]
b = [1, 3, 5, 7]

def primera_suma_multiplo_5(lista_a, lista_b):
    for x in lista_a:
        for y in lista_b:
            if (x + y) % 5 == 0:
                return x, y, x + y
    return None

resultado = primera_suma_multiplo_5(a, b)
print(resultado)

# return detiene la función completa, no solo el bucle interno
```

### 8. Alcance local, global y una alternativa preferible

**Idea clave:** Si una función asigna un valor a una variable, Python la considera local dentro de esa función. Usar 'global' modifica el estado externo, pero suele ser mejor evitarlo cuando se puede.

**Problema 8.** Corrija una función con contador usando global y luego una alternativa sin estado global.

**Qué deben saber para una pregunta de desarrollo:** identificar entradas, proceso, salida esperada y explicar por qué el código cumple la regla. Si la pregunta pide analizar código, seguir línea por línea sin adivinar.

```python
# Versión con global
contador_global = 0

def aumentar_global():
    global contador_global
    contador_global += 1
    return contador_global

print(aumentar_global())
print(aumentar_global())

# Versión preferible: pasar y retornar el nuevo estado
contador = 0

def aumentar(contador_actual):
    return contador_actual + 1

contador = aumentar(contador)
contador = aumentar(contador)
print("Sin global:", contador)
```

### 12. filter, map y reduce
**Idea clave:**
- filter(funcion, iterable) conserva los elementos que cumplen una condición.
- map(funcion, iterable) transforma cada elemento.
- reduce(funcion, iterable) acumula todo en un único resultado.

**Problema 12.** Seleccione montos aprobados, aplique impuesto y calcule total.

**Qué deben saber para una pregunta de desarrollo:** identificar entradas, proceso, salida esperada y explicar por qué el código cumple la regla. Si la pregunta pide analizar código, seguir línea por línea sin adivinar.

```python
from functools import reduce

montos = [50, 120, 200, 75, 300]

aprobados = filter(lambda m: m >= 100, montos)
con_impuesto = map(lambda m: m * 1.18, aprobados)
total = reduce(lambda a, b: a + b, con_impuesto)

print(total)

# Orden lógico:
   # 1) Filtrar: 120, 200, 300
   # 2) Transformar: 141.6, 236.0,354.0
   # 3) Reducir: 731.0
```

### 13. Generadores con yield

**Idea clave:** Una función con 'yield' no devuelve toda la lista de golpe. Produce valores uno por uno cuando se le solicita. Esto ahorra memoria y permite procesar flujos de datos.

**Problema 13.** Genere edades válidas una por una.

**Qué deben saber para una pregunta de desarrollo:** identificar entradas, proceso, salida esperada y explicar por qué el código cumple la regla. Si la pregunta pide analizar código, seguir línea por línea sin adivinar.

```python
edades = [12, 19, -5, 35, 101, 60, 17, 80]

def edades_validas(lista):
    for edad in lista:
        if 18 <= edad <= 100:
            yield edad
            
for edad in edades_validas(edades):
    print(edad)
    
# A diferencia de return [...], yield produce los valores bajo demanda.
```

### 14. Decoradores construidos desde cero, sin librerías externas

**Qué deben saber antes de resolver:**

Un decorador no es magia. Es una función que recibe otra función y devuelve una nueva función.

La estructura mínima es:

def decorador(funcion_original):

----def funcion_interna(*args, **kwargs):

--------# antes

--------resultado = funcion_original(*args, **kwargs)

--------# después

--------return resultado

----return funcion_interna

**Puntos importantes:**

funcion_original es la función que queremos envolver.
funcion_interna agrega comportamiento antes o después.
*args recibe cualquier cantidad de argumentos posicionales.
**kwargs recibe argumentos nombrados.
@decorador es una forma corta de escribir:
mi_funcion = decorador(mi_funcion).

**Pregunta de desarrollo parecida:** crear un decorador que muestre un mensaje antes y después de ejecutar una función, sin usar librerías ni funciones predefinidas especiales.

```python
def auditar(funcion_original):
    def funcion_interna(*args, **kwargs):
        print("Antes de ejecutar la función")
        resultado = funcion_original(*args, **kwargs)
        print("Después de ejecutar la función")
        return resultado
    return funcion_interna

@auditar
def calcular_importe(precio_unitario, cantidad):
    return precio_unitario * cantidad

print(calcular_importe(12.5, 4))

# Explicación:
  # 1. @auditar reemplaza calcular_importe por auditar(calcular_importe).
  # 2. Al llamar calcular_importe(12.5, 4), realmente se llama la funcion_interna.
  # 3. funcion_interna imprime el mensaje inicial.
  # 4. Luego ejecuta la funcion original con los mismos argumentos.
  # 5. Finalmente imprime el mensaje final y devuelve el resultado.
```

### 15. Decoradores profesionales: qué es functools.wraps

**Primero lo importante:** wraps no es una palabra reservada de Python. Tampoco es obligatorio para crear decoradores.

wraps viene del módulo estándar functools:

**from** functools **import** wraps

Se usa dentro de un decorador para conservar la información de la función original, por ejemplo:

- su nombre en __ name __,
- su documentación en __ doc __,
- algunos metadatos internos.

Sin *wraps*, Python puede mostrar que la función decorada se llama funcion_interna o envoltura, porque técnicamente esa es la función que reemplazó a la original.

**Qué deben saber para el examen:**

- Para entender decoradores, primero deben dominar el decorador construido desde cero.
- Si aparece @wraps(funcion), deben reconocer que solo conserva metadatos de la función original.
- @wraps(funcion) no ejecuta la función de negocio.
- La función de negocio se ejecuta cuando aparece algo como funcion(*args, **kwargs).

**Pregunta de desarrollo parecida:** crear un decorador que mida tiempo de ejecución. Aquí sí usamos wraps, pero queda explicado qué aporta y qué no aporta.

```python
import time
from functools import wraps

def medir_tiempo(funcion_original):
    @wraps(funcion_original)
    def funcion_interna(*args, **kwargs):
        inicio = time.perf_counter()
        resultado = funcion_original(*args, **kwargs)
        fin = time.perf_counter()
        print(f"Tiempo de ejecución: {fin - inicio:.6f} segundos")
        return resultado
    return funcion_interna

@medir_tiempo
def sumar_cuadrados(n):
    """Retorna la suma de cuadrados desde 0 hasta n - 1."""
    return sum(i * i for i in range(n))

print(sumar_cuadrados(10000))
print(sumar_cuadrados.__name__)
print(sumar_cuadrados.__doc__)
```

### Mini demostración: decorador sin *wraps* vs decorador con *wraps*

Este ejemplo no es un problema adicional. Es una aclaración para que no se confunda @wraps con el decorador principal.

```python
from functools import wraps

def decorador_sin_wraps(funcion):
    def interna(*args, **kwargs):
        return funcion(*args, **kwargs)
    return interna

def decorador_con_wraps(funcion):
    @wraps(funcion)
    def interna(*args, **kwargs):
        return funcion(*args, **kwargs)
    return interna

@decorador_sin_wraps
def funcion_a():
    """Documentación de función_a"""
    return "A"


@decorador_con_wraps
def funcion_b():
    """Documentación de funcion_b."""
    return "B"

print(funcion_a(), funcion_a.__name__, funcion_a.__doc__)
print(funcion_b(), funcion_b.__name__, funcion_b.__doc__)
```
